## 3. Buatlah visualisasi dengan mengunakan library python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur background visualisasi
sns.set(style="whitegrid")

# Memisahkan data berdasarkan hasil diabetes
diabetes = visual[visual['Outcome'] == 1]
no_diabetes = visual[visual['Outcome'] == 0]

# Membuat figure dan axis untuk subplot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Distribusi kadar glukosa untuk individu yang diabetes vs yang tidak diabetes
sns.histplot(diabetes['Glucose'], color='red', label='Diabetes', kde=True, ax=ax[0])
sns.histplot(no_diabetes['Glucose'], color='blue', label='Tidak Diabetes', kde=True, ax=ax[0])
ax[0].set_title('Distribusi Kadar Glukosa')
ax[0].set_xlabel('Glukosa')
ax[0].set_ylabel('Kepadatan')
ax[0].legend()

# Plot 2: Scatter plot antara BMI dan kadar glukosa dengan warna berdasarkan hasil diabetes
sns.scatterplot(data=visual, x='BMI', y='Glucose', hue='Outcome', palette='coolwarm', ax=ax[1])
ax[1].set_title('BMI vs Kadar Glukosa')
ax[1].set_xlabel('BMI')
ax[1].set_ylabel('Glukosa')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

1.   Distribusi Kadar Glukosa: Grafik pertama menunjukkan distribusi kadar glukosa pada pasien dengan diabetes dan tanpa diabetes. Untuk pasien dengan diabetes, nilai glukosa lebih tinggi (sekitar 150 hingga 200), sedangkan untuk pasien tanpa diabetes, nilai glukosa umumnya lebih rendah.
2.   Hubungan antara BMI dan Kadar Glukosa: Grafik kedua menunjukkan plot dispersi antara BMI dan glukosa, dengan warna yang membedakan pasien diabetes dan non-diabetes. Namun, tidak ada hubungan langsung yang terlihat antara BMI dan glukosa, tetapi pasien dengan kadar glukosa yang lebih tinggi cenderung memiliki BMI yang lebih tinggi.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
visual = pd.read_csv('/content/diabetes_preprocessed.csv')

# Menghitung jumlah pasien dengan dan tanpa diabetes
outcome_counts = visual['Outcome'].value_counts()

# Mendefinisikan label dan ukuran untuk pie chart
labels = ['Tidak Diabetes (0)', 'Diabetes (1)']
sizes = outcome_counts.values
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # Memisahkan irisan pertama (Tidak Diabetes)

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Proporsi Hasil Diabetes')
plt.axis('equal')  # Rasio aspek yang sama memastikan pie digambar sebagai lingkaran.
plt.show()
```


Penjelasan:  
*   Pie chart ini akan menunjukkan proporsi pasien dalam dataset dengan dan tanpa diabetes.
*   Tanda "0" menunjukkan tidak memiliki diabetes dan "1" menunjukkan diabetes.
*   Sizes dihitung dari total hasil yang dihitung dari dataset.
*   Warna memberikan warna yang berbeda untuk kedua kategori, dan explode digunakan untuk menonjolkan bagian "No Diabetes."

Interpretasi: Pie chart ini akan membantu memahami distribusi data dengan memberikan gambaran cepat tentang jumlah pasien dengan diabetes dibandingkan dengan yang tidak.
